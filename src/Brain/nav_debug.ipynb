```python
import math
```

#### for i in range(0, 181)로 설정한 이유: ####

orientation 좌표계는 좌표평면 좌측 축 0도 기준으로 시계방향으로 한 바퀴 회전하면서 cos 0도부터 180도까지 표현되는 좌표계입니다.

일반적인 극좌표계를 a, orientation 좌표계를 b라고 하자면,

a 기준에서 보면

0도(a) = 90도(b),
90도(a) = 45도(b),
180도(a) = 0도(b), 180도(a)
270도(a) = 135도(b),
360도(a) = 90도(b)

b 기준에서 보면

0도(b) = 180도(a)
45도(b) = 90도(a)
90도(b) = 0도(a), 360도(a)
135도(b) = 270도(a)
180도(b) = 180도(a)

이를 0도부터 360도의 구간으로 확장하는 변환 코드가 converted_ori_cos_radian

```python
for i in range(0, 181):
    orientation_cos = math.cos(math.radians(i))

    curr_x, curr_y = (120, 200)
                
    tar_x, tar_y = (180, 260)

    x_diff = tar_x - curr_x
    y_diff = -(tar_y - curr_y)

    """내 위치에서 waypoint까지의 각도"""
    target_radian = math.atan2(y_diff, x_diff)
    target_degree = math.degrees(target_radian)

    """내가 바라보는 방향"""
    converted_ori_cos_radian = orientation_cos
    converted_ori_cos_degree = math.degrees(math.acos(converted_ori_cos_radian))
    converted_ori_cos_degree *= 2
    converted_ori_cos = (180 - converted_ori_cos_degree)% 360

    delta_degree = abs(target_degree - converted_ori_cos)

    clockwise, revert_clockwise = delta_degree, 360 - delta_degree

    if i % 10 == 0:
        print("===============")
        print(f"내가 바라보는 방향: {converted_ori_cos}'")
        print(f"목표 방향: {target_degree:.3f}\n")

        print(f"시계방향으로 {clockwise:.3f}도만큼 회전해야 한다.")
        print(f"반시계방향으로 {revert_clockwise:.3f}도만큼 회전해야 한다.\n")

        if clockwise < revert_clockwise:
            print("우회전") # 우회전
        else:
            print("좌회전") # 좌회전
```
