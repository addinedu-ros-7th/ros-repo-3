#### 코드 설명 ####

orientation 좌표계는 좌표평면 좌측 축 0도 기준으로 시계방향으로 한 바퀴 회전하면서 cos 0도부터 180도까지 표현되는 좌표계입니다.

일반적인 극좌표계를 a, orientation 좌표계를 b라고 둡니다.

##### a 기준에서 보면 

0도(a) = 90도(b),   
90도(a) = 45도(b),   
180도(a) = 0도(b), 180도(b)   
270도(a) = 135도(b),   
360도(a) = 90도(b)   

##### b 기준에서 보면

0도(b) = 180도(a),   
45도(b) = 90도(a),   
90도(b) = 0도(a), 360도(a),   
135도(b) = 270도(a),   
180도(b) = 180도(a),   

for 반복문의 범위가 0~180인 이유가 여기 있습니다.
b는 0도부터 180도 사이의 주기를 가지기 때문입니다.


이 코드는 b좌표계에 주어지는 orientation.z 값들을 a좌표계로 변환하고, a좌표계 위에 주어진 waypoint와 curr_pose를 활용하여 이 두 기울기 차이를 구해 시계방향(우회전), 반시계방향(좌회전) 회전 여부를 결정하는 코드입니다.

```python
import matplotlib.pyplot as plt
```

```python
for i in range(0, 181):
    orientation_cos = math.cos(math.radians(i))

    curr_x, curr_y = (300, 300)
                
    tar_x, tar_y = (100, 100)

    x_diff = tar_x - curr_x
    y_diff = -(tar_y - curr_y) # 맵이 y축 기준 상하반전이므로.

    """내 위치에서 waypoint까지의 각도"""
    target_radian = math.atan2(y_diff, x_diff)
    target_degree = math.degrees(target_radian) % 360

    """내가 바라보는 방향"""
    converted_ori_cos_radian = orientation_cos
    converted_ori_cos_degree = math.degrees(math.acos(converted_ori_cos_radian))
    converted_ori_cos_degree *= 2
    converted_ori_cos = (180 - converted_ori_cos_degree)% 360

    delta_degree = target_degree - converted_ori_cos

    if delta_degree > 0:
        clockwise = 360 - abs(delta_degree)
        revert_clockwise = 360 - clockwise
    elif delta_degree < 0:
        clockwise = abs(delta_degree)
        revert_clockwise = 360 - clockwise

    if i % 10 == 0:
        print("===============")
        print(f"내가 바라보는 방향: {converted_ori_cos:.2f}'")
        print(f"목표 방향: {target_degree:.2f}\n")

        print(f"시계방향으로 {clockwise:.2f}도만큼 회전해야 한다.")
        print(f"반시계방향으로 {revert_clockwise:.2f}도만큼 회전해야 한다.\n")

        if clockwise < revert_clockwise:
            print("우회전") # 우회전
        else:
            print("좌회전") # 좌회전
```

```python
import math
import matplotlib.pyplot as plt
import numpy as np

def plot_position(curr_x, curr_y, tar_x, tar_y, orientation_degree):
    plt.figure(figsize=(6, 6))
    plt.xlim(0, 400)
    plt.ylim(0, 400)
    
    # 현재 위치
    plt.plot(curr_x, curr_y, 'bo', markersize=10, label='curr_pos')
    
    # 목표 위치
    plt.plot(tar_x, tar_y, 'r*', markersize=12, label='way_point')
    
    # 내가 바라보는 방향 벡터
    arrow_length = 50  # 화살표 길이
    radian = math.radians(orientation_degree)
    arrow_dx = arrow_length * math.cos(radian)
    arrow_dy = arrow_length * math.sin(radian)

    plt.arrow(curr_x, curr_y, arrow_dx, arrow_dy, head_width=10, head_length=10, fc='g', ec='g', label='my_view')
    
    # 목표까지의 직선
    plt.plot([curr_x, tar_x], [curr_y, tar_y], 'k--', label='line to waypoint')
    
    plt.legend()
    plt.title(f'curr_view: {orientation_degree:.2f}°')
    plt.grid()
    plt.show()



curr_x, curr_y = (300, 300)
tar_x, tar_y = (100, 100)

for i in range(0, 181, 20):  # 30도 간격으로 시각화
    orientation_cos = math.cos(math.radians(i))
    
    # 목표 방향 계산
    x_diff = tar_x - curr_x
    y_diff = -(tar_y - curr_y)  # y축 반전 고려
    target_radian = math.atan2(y_diff, x_diff)
    target_degree = math.degrees(target_radian) % 360
    
    # 바라보는 방향 계산
    converted_ori_cos_radian = orientation_cos
    converted_ori_cos_degree = math.degrees(math.acos(converted_ori_cos_radian))
    converted_ori_cos_degree *= 2
    converted_ori_cos = (180 - converted_ori_cos_degree) % 360

    delta_degree = target_degree - converted_ori_cos

    print(f"내가 바라보는 방향: {converted_ori_cos:.2f}'")
    print(f"목표 방향: {target_degree:.2f}\n")
    
    if delta_degree > 0:
        clockwise = 360 - abs(delta_degree)
        revert_clockwise = 360 - clockwise
    elif delta_degree < 0:
        clockwise = abs(delta_degree)
        revert_clockwise = 360 - clockwise

    print(f"시계방향으로 {clockwise:.2f}도만큼 회전해야 한다.")
    print(f"반시계방향으로 {revert_clockwise:.2f}도만큼 회전해야 한다.\n")

    print(converted_ori_cos)
    
    plot_position(curr_x, curr_y, tar_x, tar_y, converted_ori_cos)
```
